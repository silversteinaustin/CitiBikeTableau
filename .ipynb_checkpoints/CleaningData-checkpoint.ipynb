```python
# Import dependencies
import pandas as pd
from pathlib import Path
import os
```

```python
#Path to CSV file
citibike_Jan2014 = Path(r"C:\Users\silve\CitiBikeTableau\201401-citibike-tripdata.csv")

# Read our Jan 2014 csv files into pandas dataframes
citibike_Jan2014_df = pd.read_csv(citibike_Jan2014)
citibike_Jan2014_df.head()
```

```python
#Path to CSV file
citibike_Jan2015 = Path(r"C:\Users\silve\CitiBikeTableau\201501-citibike-tripdata.csv")

# Read our Jan 2015 csv files into pandas dataframes
citibike_Jan2015_df = pd.read_csv(citibike_Jan2015)
citibike_Jan2015_df.head()
```

```python
#Drop rows that have missing data
cleaned_Jan2014_df = citibike_Jan2014_df.dropna()
cleaned_Jan2014_df.head()
```

```python
#Drop rows that have missing data
cleaned_Jan2015_df = citibike_Jan2015_df.dropna()
cleaned_Jan2015_df.head()
```

```python
# Quick overview of the DataFrame's structure and to quickly assess data types and missing values
cleaned_Jan2014_df.info()
cleaned_Jan2015_df.info()
```

```python
num_rows, num_columns = cleaned_Jan2014_df.shape

# Print the number of rows and columns in the DataFrame
print(f"Number of rows: {num_rows} in the January 2014 DataFrame")
print(f"Number of columns: {num_columns} in the January 2014 DataFrame")
```

```python
num_rows, num_columns = cleaned_Jan2015_df.shape

# Print the number of rows and columns in the DataFrame
print(f"Number of rows: {num_rows} in the January 2015 DataFrame")
print(f"Number of columns: {num_columns} in the January 2015 DataFrame")
```
